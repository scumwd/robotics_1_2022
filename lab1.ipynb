```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 40, density=True)
plt.plot(bins, 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu) ** 2 / (2 * sigma ** 2)), linewidth=2, color='r')
plt.show()
```

```python
import numpy as np
import seaborn as sns

n = 100000
x1 = np.random.normal(-5, 2, int(n * 0.5))
x2 = np.random.normal(5, 3, int(n * 0.5))
x = np.concatenate((x1, x2))
np.random.shuffle(x)

samples = 1000
sample_value = np.random.rand(samples)
for i in range(0,samples):
    j = np.random.randint(1,n,10)
    sample_value[i] = x[j].mean()

plt.subplot(1,2,1)
sns.histplot(sample_value, kde = False)
sns.kdeplot(x)
plt.title('Histogram sample value')
plt.xlabel('Sample value')
plt.ylabel('Count')
plt.subplot(1,2,2)
sns.histplot(sample_value, kde = True)
plt.title('Density sample value')
plt.xlabel('Sample value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.arange(-5, 5, 0.1)
y = (4 * x**3) - (10 * x**2) + (2*x) + 1
noise = 80 * np.random.normal(size=x.size)
ydata = y + noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.show()
```

```python
def func(x, a, b, c, d):
    return a * 4 * x**3  - b* 10 * x**2 + c*2*x + d + 1

coeff, cov = optimize.curve_fit(func,x, ydata)
print(coeff)

plt.plot(x, func(x, coeff[0], coeff[1], coeff[2], coeff[3]), 'ro', label='curve fit result')
plt.plot(x, ydata, 'bo', label='scattered data with noise')
plt.plot(x, y, 'g', label='data gotten using formula')
plt.legend()
plt.show()
```
