```python
from sklearn.linear_model import RANSACRegressor
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.arange(0, N, 1)
x2d = np.column_stack((x,))
y = 2*x + 7

gauss = np.random.normal(0,50,N)
ynoise = y + gauss

plt.plot(x, ynoise, 'bo')
plt.plot(x, y, 'r')

plt.show()
```

```python
ransac = RANSACRegressor()

ransac.fit(x2d , ynoise)
plt.figure(figsize=(15, 7))
# ploting the training dataset in scattered graph
plt.scatter(x2d, ynoise, color='red')

# ploting the testing dataset in line line
plt.plot(x2d, ransac.predict(x2d), color='green')

plt.title('Hours vs Score')
plt.xlabel('Hours')
plt.ylabel('Score')

plt.show()
```
