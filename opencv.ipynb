```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

imgPath = 'scumwd.jpg'
image = cv2.imread(imgPath)
size = image.shape
print(size)
```

```python
columns = 4
rows = 4
dividers = (int(size[1]/columns), int(size[0]/rows))
```

```python
for i in range(0, rows):
    for j in range(0, columns):
        dx = dividers[0]
        dy = dividers[1]
        hx0 = dx * j
        hx1 = dx * (j + 1)
        hy0 = dy * i
        hy1 = dy * (i + 1)
        img = image[hy0:hy1, hx0:hx1]
        cv2.imwrite(f"scumwd({i}-{j}).jpg", img)
```

```python
fig, ax = plt.subplots(nrows=rows, ncols=columns)
for i in range(0, rows):
    for j in range(0, columns):
        img = plt.imread(f"scumwd({i}-{j}).jpg")
        ax[i,j].imshow(img)
        ax[i,j].axis('off')
plt.tight_layout()
plt.show()
```
